# Pairwise Distance

- Get pairwise distance to all samples
- Repopulate with metadata
- Look at top N samples and compare

```python
from sklearn.metrics import pairwise_distances
import numpy as np
```

```python
def pairwise_distance_ranks(sample, df, genes, group):
    dist = pairwise_distances(np.array(sample[genes]).reshape(1, -1), df[genes])
    dist = pd.DataFrame([dist.ravel(), df[group]]).T
    dist.columns = ['Distance', 'Class']

    # Pandas-FU
    med_dist = dist.groupby('Class').apply(lambda x: x['Distance'].median()).reset_index()
    med_dist.columns = ['Class', 'MedianDistance']
    return med_dist.sort_values('MedianDistance').reset_index(drop=True)
```

## Change number of embeddings

```python
def umap_distances(sample, df, genes, group, n_components=2):
    """
    Run UMAP to create an embedding of the sample and background dataset, then calculate distance to each
    group via pairwise distance

    Args:
        sample: n-of-1 sample. Gets own label
        df: background dataset
        genes: genes to use for pairwise distance
        group: Column to use as class discriminator

    Returns:
        DataFrame of UMAP pairwise distances
    """
    # Concatenate sample to background
    concat = df.append(sample)
    concat.tail(2)

    # UMAP embedding
    cols = [f'UMAP{i+1}' for i in range(n_components)]
    embedding = pd.DataFrame(umap.UMAP(n_components=n_components).fit_transform(concat[genes]), 
                             columns=cols)
    embedding[group] = concat[group].values

    # Separate sample and background
    sample = embedding.iloc[-1]
    df = embedding.iloc[:-1]

    # Pairwise distances
    dist = pairwise_distances(np.array(sample[cols]).reshape(1, -1), df[cols])
    dist = pd.DataFrame([dist.ravel(), df[group]]).T
    dist.columns = ['Distance', 'Group']

    # Median by group and sort
    med_dist = dist.groupby('Group').apply(lambda x: x['Distance'].median()).reset_index()
    med_dist.columns = ['Group', 'MedianDistance']
    return med_dist.sort_values('MedianDistance').reset_index(drop=True)
```

# PCA

if `0 < n_components < 1` and `svd_solver == âfullâ`, select the number of components such that the amount of variance that needs to be explained is greater than the percentage specified by n_components

```python
from sklearn.decomposition import PCA
from sklearn.metrics import pairwise_distances
```

## GTEx
Identify inflection point in GTEx

```python
pca = PCA().fit(gtex[genes])
```

```python
plt.figure(figsize=(6, 4))
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.xlim([0, 250])
plt.axvline(25, c='r', label='Inflection Point')
plt.ylabel('Cumulative Explained Variance')
plt.title('PCA Explained Variance and Inflection Point');
```

```python
plt.figure(figsize=(6, 4))
plt.plot(pca.explained_variance_ratio_)
plt.xlabel('Number of Components')
plt.xlim([0, 250])
plt.axvline(25, c='r', label='Inflection Point')
plt.ylabel('Explained Variance')
plt.title('PCA Explained Variance and Inflection Point');
```

```python
def pca_distances(sample, df, genes, group, n_components=50):
    # Concatenate sample to background
    concat = df.append(sample)

    # PCA embedding
    embedding = pd.DataFrame(PCA(n_components=n_components).fit_transform(concat[genes]))
    embedding[group] = concat[group].values
    components = list(embedding.columns[:-1])

    # Separate sample and background
    sample = embedding.iloc[-1]
    df = embedding.iloc[:-1]

    # Pairwise distances
    dist = pairwise_distances(np.array(sample[components]).reshape(1, -1), df[components])
    dist = pd.DataFrame([dist.ravel(), df[group]]).T
    dist.columns = ['Distance', 'Group']

    # Median by group and sort
    med_dist = dist.groupby('Group').apply(lambda x: x['Distance'].median()).reset_index()
    med_dist.columns = ['Group', 'MedianDistance']
    return med_dist.sort_values('MedianDistance').reset_index(drop=True)
```
