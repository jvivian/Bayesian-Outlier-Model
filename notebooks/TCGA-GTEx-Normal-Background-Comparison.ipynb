# Compare All Background Datasets

# Inputs

```python
import warnings
warnings.filterwarnings('ignore')

import rnaseq_lib3 as r

import pandas as pd
import os
import scipy.stats as st
import pymc3 as pm

import seaborn as sns
import matplotlib.pyplot as plt
import holoviews as hv
hv.extension('bokeh')
```

```python
# Read in expression data
df = pd.read_hdf('/mnt/data/expression/tcga_gtex_tpm_norm_filt.hd5')
# Subset
gtex = df[df.label == 'gtex'].sort_values('tissue')
normal = df[df.label == 'tcga-normal'].sort_values('tissue')
tumor = df[df.label == 'tcga-tumor'].sort_values('tissue')
# Read in drug genes
genes = df.columns[5:]
drug_genes = [x.split('\t')[0] for x in open('../data/druggable-genes.tsv', 'r').readlines()]
drug_genes = [x for x in drug_genes if x in df.columns]
```

```python
sample_id = 'TCGA-FY-A3I4-01'
sample = tumor.loc[sample_id]
sample.subtype
```

```python
out_dir = '/mnt/research_serializations/TCGA-GTEx-Normal-Background-Comparison'
```

# Run with All Samples

Add column that uniquely identifies tissue and dataset so tissues don't get merged between datasets when calculating distance. 

```python
col = df.apply(lambda x: f'{x.label}-{x.tissue}', axis=1)
df.insert(0, 'ds-label', col)
```

```python
ranks = r.outlier.pairwise_distance_ranks(sample, df.drop(sample_id), genes, 'ds-label')
training_group = df[df['ds-label'].isin(ranks.head(5).Group)]
    
training_genes = r.outlier.select_k_best_genes(training_group, genes, 'ds-label', 40)
training_genes = list(set(drug_genes + training_genes))
```

```python
out = os.path.join(out_dir, 'all-samples.pkl')
if os.path.exists(out):
    m, t = r.outlier.load_model(out)
else:
    m, t = r.outlier.run_model(sample, training_group, training_genes, 'ds-label')
    r.outlier.pickle_model(out, m, t)
```

## Traceplot

```python
pm.traceplot(t, varnames=['a', 'b', 'eps']);
```

## Weights

```python
r.outlier.plot_weights(training_group['ds-label'].unique(), t)
```

## Pvals

```python
ppp = {}
ppc = r.outlier.ppc(t, drug_genes)
ppp['all'] = r.outlier.posterior_predictive_pvals(sample, ppc)
```

## Compare TCGA / GTEx / Normal P-values

```python
ppp = pd.DataFrame(ppp)
```

### GTEx vs. Normal

```python
print(st.pearsonr(ppp['gtex'], ppp['normal']))
r.plot.regression(ppp, 'gtex', 'normal', title='P-values Between GTEx and Normal')
```

### GTEx vs. Tumor

```python
print(st.pearsonr(ppp['gtex'], ppp['tumor']))
r.plot.regression(ppp, 'tumor', 'gtex', title='P-values Between GTEx and Tumor')
```

### Tumor vs. Normal

```python
print(st.pearsonr(ppp['normal'], ppp['tumor']))
r.plot.regression(ppp, 'tumor', 'normal', title='P-values Between Normal and Tumor')
```

## Distributions

```python
plt.figure(figsize=(12, 4))
plt.axvline(0.05, c='r', label='Sig Cutoff')
plt.axvline(0.95, c='r')
sns.kdeplot(ppp['tumor'], label='Tumor', shade=True, alpha=0.25, color='orange')
sns.kdeplot(ppp['gtex'], label='GTEx', shade=True, alpha=0.25, color='green')
sns.kdeplot(ppp['normal'], label='Normal', shade=True, alpha=0.25, color='blue')
plt.xlim([0, 1])
plt.xlabl()
plt.legend(bbox_to_anchor=(0.8, 0.95))
plt.title('P-value Distribution Given Background Dataset');
```
