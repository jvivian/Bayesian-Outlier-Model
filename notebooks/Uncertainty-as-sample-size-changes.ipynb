```python
# Read in centered data
df = pd.read_hdf('/mnt/data/expression/tcga_gtex_tpm_norm_filt.hd5')
# Subset
gtex = df[df.label == 'gtex'].sort_values('tissue')
normal = df[df.label == 'tcga-normal'].sort_values('tissue')
tumor = df[df.label == 'tcga-tumor'].sort_values('tissue')
# Read in drug genes
genes = df.columns[5:]
drug_genes = [x.split('\t')[0] for x in open('../data/druggable-genes.tsv', 'r').readlines()]
drug_genes = [x for x in drug_genes if x in df.columns]
```

```python
out_dir = '/mnt/research_serializations/Uncertainty-as-sample-size-changes'
thyroid_samples = list(gtex[gtex.tissue == 'Thyroid'].index)
num_samples = [0, 1, 5, 10, 20, 50, 100, 278]
for i in tqdm(num_samples):
    sub = gtex[gtex.tissue != 'Thyroid']
    samples = np.random.choice(thyroid_samples, i, replace=False)
    sub = pd.concat([sub, gtex.loc[samples]])
    sub.to_hdf(os.path.join(out_dir, f'bg-thyroid-{i}.hd5'), key='exp')
```

# Perform Runs

Run the above using the `outlier-detection` command line tool

```python
def run(background):
    out = os.path.join(out_dir, 'output', os.path.basename(background))
    out = os.path.splitext(out)[0]
    os.makedirs(out)
    parameters = ['-s', '/mnt/data/expression/tumor.hd5',
                  '-b', background,
                  '--name', 'TCGA-DJ-A2PX-01',
                  '-l', '/mnt/data/outlier/drug-genes.txt',
                  '-o', out,
                  '-c', '5']
    subprocess.check_call(['outlier-detection'] + parameters)
    print(' '.join(parameters))
```

```python
backgrounds = [os.path.join(out_dir, f'bg-thyroid-{i}.hd5') for i in num_samples][2:]
```

```python
for bg in backgrounds:
    run(bg)
    break
```

## P-value Correlation

```python
out_dir = '/mnt/research_serializations/Uncertainty-as-sample-size-changes/output'
pvals = []
for run in os.listdir(out_dir):
    pvals.append(
        pd.read_csv(os.path.join(out_dir, run, 'TCGA-DJ-A2PX-01', 'pvals.tsv'), sep='\t', index_col=0)
    )
pvals = pd.concat(pvals, axis=1).dropna()
pvals.columns = [1, 5, 10, 20, 50, 100, 278]
```

```python
from itertools import combinations
```

```python
for i in combinations(pvals.columns, 2):
    print(i)
    break
```

```python
f, ax = plt.subplots(3, 7, figsize=(12, 12), sharex=True, sharey=True)
ax = ax.flatten()
for i, (x, y) in enumerate(combinations(pvals.columns, 2)):
    sns.regplot(pvals[x], pvals[y], ax=ax[i])
    r, _ = st.pearsonr(pvals[x], pvals[y])
    ax[i].set_title(f'R: {round(r, 2)}')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.tight_layout()
```
