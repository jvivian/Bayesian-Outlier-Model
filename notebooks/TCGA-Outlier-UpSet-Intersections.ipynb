# TCGA Outlier Intersection Plots

Collect and visualize outlier intersections between the following groups:

- PanTumor
- PanNormal
- PanGTEx
- MatchedTumor
- MatchedNormal
- MatchedGTEx
- Bayesian Outlier Model

# Inputs

```python
import warnings
warnings.filterwarnings('ignore')

import rnaseq_lib3 as r

import pandas as pd
import os
import scipy.stats as st
import pymc3 as pm
from tqdm import tqdm_notebook as tqdm

import seaborn as sns
import matplotlib.pyplot as plt
import holoviews as hv
hv.extension('bokeh')
```

```python
# Read in expression data
df = pd.read_hdf('/mnt/data/expression/tcga_gtex_tpm_norm_filt.hd5')
# Subset
gtex = df[df.label == 'gtex'].sort_values('tissue')
normal = df[df.label == 'tcga-normal'].sort_values('tissue')
tumor = df[df.label == 'tcga-tumor'].sort_values('tissue')
# Read in drug genes
genes = df.columns[5:]
drug_genes = [x.split('\t')[0] for x in open('../data/druggable-genes.tsv', 'r').readlines()]
drug_genes = [x for x in drug_genes if x in df.columns]
```

```python
sample_dir = '/mnt/outlier-runs/tcga-run/'

tumor_sub = tumor[['id', 'tissue'] + drug_genes]
samples = os.listdir(sample_dir)
samples = [tumor_sub.loc[x] for x in samples]
```

# Calculate Outliers
For each type, store `sample-gene`

```python
from collections import defaultdict
outliers = defaultdict(list)
```

## PanTumor/PanNormal/PanGTEx

```python
for gene in tqdm(drug_genes):
    _, tumor_upper = r.math.iqr_bounds(tumor[gene])
    _, normal_upper = r.math.iqr_bounds(normal[gene])
    _, gtex_upper = r.math.iqr_bounds(gtex[gene])
    for sample in samples:
        value = sample[gene]
        if value > tumor_upper:
            outliers['PanTumor'].append(f'{sample.id}-{gene}')
        if value > normal_upper:
            outliers['PanNormal'].append(f'{sample.id}-{gene}')
        if value > gtex_upper:
            outliers['PanGTEx'].append(f'{sample.id}-{gene}')
```

## MatchedTumor/MatchedNormal/MatchedGTEx

```python
tumor_sub = tumor[['tissue'] + drug_genes]
normal_sub = normal[['tissue'] + drug_genes]
gtex_sub = gtex[['tissue'] + drug_genes]

for sample in tqdm(samples):
    t_sub = tumor_sub[tumor_sub.tissue == sample.tissue]
    n_sub = normal_sub[normal_sub.tissue == sample.tissue]
    g_sub = gtex_sub[gtex_sub.tissue == sample.tissue]
    for gene in drug_genes:
        _, tumor_upper = r.math.iqr_bounds(t_sub[gene])
        _, normal_upper = r.math.iqr_bounds(n_sub[gene])
        _, gtex_upper = r.math.iqr_bounds(g_sub[gene])
        
        value = sample[gene]
        if value > tumor_upper:
            outliers['MatchedTumor'].append(f'{sample.id}-{gene}')
        if value > normal_upper:
            outliers['MatchedNormal'].append(f'{sample.id}-{gene}')
        if value > gtex_upper:
            outliers['MatchedGTEx'].append(f'{sample.id}-{gene}')
```

## Bayesian Outlier Model

```python
for sample in tqdm(os.listdir(sample_dir)):
    pvals = pd.read_csv(os.path.join(sample_dir, sample, 'pvals.tsv'), sep='\t', index_col=0, header=None)
    pvals.columns = ['pval']
    pvals = pvals.loc[drug_genes]
    for gene in pvals[pvals.pval <= 0.05].index:
        outliers['Model'].append(f'{sample}-{gene}')
```

# Calculate Mutually Exclusive Set Intersections

```python
for key in outliers:
    outliers[key] = set(outliers[key])
```

```python
from itertools import combinations

def mutually_exclusive_set_counts(sets, groups):
    # Collect master set
    master = set()
    for g in groups:
        for s in sets[g]:
            master.add(s)
    
    combs = []
    counts = []
    # For decreasing set sizes, identify if outlier belongs to group
    for i in range(len(groups), 0, -1):
        for combination in combinations(groups, i):
            count = 0
            combs.append(combination)
            inter = set.intersection(*[sets[x] for x in combination])
            
            # Iterate over master and pop items if they belong to group
            for item in list(master):
                if item in inter:
                    count += 1
                    master.remove(item)

            counts.append(count)
    return combs, counts
```

# UpSet Plots

```python
import upsetplot
```

## All Categories

```python
comb, counts = mutually_exclusive_set_counts(outliers, outliers.keys())
upset = upsetplot.from_memberships(comb, data=counts)
upsetplot.plot(upset, sort_by='cardinality');
```

## Pan-Methods and Model

```python
comb, counts = mutually_exclusive_set_counts(outliers, ['PanTumor', 'PanGTEx', 'PanNormal', 'Model'])
upset = upsetplot.from_memberships(comb, data=counts)
upsetplot.plot(upset, sort_by='cardinality', show_counts=True)
plt.title('Outlier Intersections by Method');
```

## Matched-Methods and Model

```python
comb, counts = mutually_exclusive_set_counts(
    outliers, ['MatchedTumor', 'MatchedGTEx', 'MatchedNormal', 'Model']
)
upset = upsetplot.from_memberships(comb, data=counts)
upsetplot.plot(upset, sort_by='cardinality', show_counts=True)
plt.title('Outlier Intersections by Method');
```
