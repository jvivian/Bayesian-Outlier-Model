# Bayesian Outlier N-of-1 Example

A series of investigative plots and pval estimates for an n-of-1 sample

1. Trimap of n-of-1 within context of normal dataset
    - Inputs
        - N-of-1 `pd.Series`
        - Background dataset `pd.DataFrame`
    - Outputs
        - Interactive Holoviews plot of sample within the context of the background dataset.

2. Treehouse Method for Comparison
    - Inputs
        - N-of-1 `pd.Series`
        - Background Dataset `pd.DataFrame`
        - Matching Dataset (derived from above)
    - Outputs
        - 2 plots of background distribution, n-of-1, and cutoff.

3. Experiment Subset
    - Inputs
        - N-of-1 `pd.Series`
        - Background dataset `pd.DataFrame`
        - Matching dataset (derived from above)
        - Match-Minus dataset (derived)
        - Match-subsets (derived)
        - Match-Minus + subsets (derived)
    - Ouputs
        - Weight plot (if applicable)
        - Pval estimates

# Inputs

```python
import rnaseq_lib3 as r

import pymc3 as pm
import pandas as pd
import numpy as np
from scipy.stats import pearsonr

import matplotlib.pyplot as plt
import seaborn as sns
import holoviews as hv
hv.extension('bokeh')

import warnings
warnings.filterwarnings('ignore')
```

```python
# Read in centered data
df = pd.read_hdf('/mnt/data/expression/tcga_gtex_tpm_norm_filt_centered.hd5')
# Subset
gtex = df[df.label == 'gtex'].sort_values('tissue')
normal = df[df.label == 'tcga-normal'].sort_values('tissue')
tumor = df[df.label == 'tcga-tumor'].sort_values('tissue')
# Read in drug genes
genes = df.columns[5:]
drug_genes = [x.split('\t')[0] for x in open('../data/druggable-genes.tsv', 'r').readlines()]
drug_genes = [x for x in drug_genes if x in df.columns]
```

```python
sample = tumor.loc['TCGA-KK-A7B2-01']
classes = normal.tissue.unique()
```

# Sample Plots

## Tumor

```python
plt.figure(figsize=(12, 4))
sns.countplot(x='tissue', data=tumor, order=tumor.tissue.value_counts().index)
plt.xticks(rotation=90)
plt.title("TCGA-Tumor Counts by Tissue");
```

## Normal

```python
plt.figure(figsize=(12, 4))
sns.countplot(x='tissue', data=normal, order=normal.tissue.value_counts().index)
plt.xticks(rotation=90)
plt.title("TCGA-Normal Counts by Tissue");
```

## GTEx

```python
plt.figure(figsize=(12, 4))
sns.countplot(x='tissue', data=gtex, order=gtex.tissue.value_counts().index)
plt.xticks(rotation=90)
plt.title("GTEx Counts by Tissue");
```

# Genes of Interest

```python
f, ax = plt.subplots(1, 2, figsize=(12, 4))
for i, gene in enumerate(['ARAF', 'PARP1']):
    ax[i].axvline(sample[gene], c='r', label='N-of-1')
    sns.kdeplot(normal[gene], ax=ax[i], shade=True, alpha=0.5, legend=False)
    ax[i].set_title(gene)
```

# IQR Cutoff (Treehouse Method)

## All Normals

```python
f, ax = plt.subplots(1, 2, figsize=(12, 4))
for i, gene in enumerate(['ARAF', 'PARP1']):
    _, upper = r.math.iqr_bounds(normal[gene])
    ax[i].axvline(upper, c='g', label='IQR-Cutoff')
    ax[i].axvline(sample[gene], c='r', label='N-of-1')
    sns.kdeplot(normal[gene], ax=ax[i], shade=True, alpha=0.5)
```

## Matching Tissue (Prostate)

```python
matched_normal = normal[normal.tissue == sample.tissue]
```

```python
f, ax = plt.subplots(1, 2, figsize=(12, 4))
for i, gene in enumerate(['ARAF', 'PARP1']):
    _, upper = r.math.iqr_bounds(matched_normal[gene])
    ax[i].axvline(upper, c='g', label='IQR-Cutoff')
    ax[i].axvline(sample[gene], c='r', label='N-of-1')
    sns.kdeplot(matched_normal[gene], ax=ax[i], shade=True, alpha=0.5)
```

# Remove Matched Tissue

```python
other_normals = normal[normal.tissue != sample.tissue]
```

## TriMap

```python
on_trimap = r.outlier.dimensionality_reduction(sample, other_normals, genes=genes, col='tissue')
```

```python
%%opts Scatter [width=800 height=600 color_index='tissue' legend_position='left']
%%opts Scatter [tools=[hover] size_index='size']
%%opts Scatter (cmap='Category20' size=5 alpha=0.5)
on_trimap
```

## TSNE

```python
on_tsne = r.outlier.dimensionality_reduction(sample, other_normals, genes=genes,
                                               col='tissue', method='tsne')
```

```python
%%opts Scatter [width=800 height=600 color_index='tissue' legend_position='left']
%%opts Scatter [tools=[hover] size_index='size']
%%opts Scatter (cmap='Category20' size=5 alpha=0.5)
on_tsne
```

## IQR Cutoff of Closest Matching Tissue

```python
breast = normal[normal.tissue == 'Breast']
```

```python
f, ax = plt.subplots(1, 2, figsize=(12, 4))
for i, gene in enumerate(['ARAF', 'PARP1']):
    _, upper = r.math.iqr_bounds(breast[gene])
    ax[i].axvline(upper, c='g', label='IQR-Cutoff')
    ax[i].axvline(sample[gene], c='r', label='N-of-1')
    sns.kdeplot(breast[gene], ax=ax[i], shade=True, alpha=0.5)
```

## Bayesian Model

```python
n = 100
training_genes = r.outlier.select_k_best_genes(normal, genes, 'tissue', n=n)
```

```python
m_other, t_other = r.outlier.run_model(sample, other_normals, 'tissue', 
                                       training_genes=training_genes, 
                                       beta_func=pm.LogitNormal)
```

### Traceplot

```python
pm.traceplot(t_other);
```

### Model Weights

```python
r.outlier.plot_weights(other_normals.tissue.unique(), t_other)
```

### Posterior Predictive Checks

```python
ppc_other = r.outlier.ppc_from_coefs(t_other, drug_genes, other_normals, 'tissue')
```

```python
pvals_other = r.outlier.posterior_predictive_pvals(sample, ppc_other)
```

```python
f, ax = plt.subplots(1, 2, figsize=(12, 4))
for i, gene in enumerate(['ARAF', 'PARP1']):
    r.outlier.plot_gene_ppc(sample, ppc_other, gene, ax=ax[i])
```
