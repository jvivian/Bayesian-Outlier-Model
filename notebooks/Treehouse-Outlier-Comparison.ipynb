## Inputs

```python
import warnings 
warnings.filterwarnings('ignore')

import rnaseq_lib3 as r
import pandas as pd
import numpy as np
import os

import matplotlib.pyplot as plt
import seaborn as sns
import holoviews as hv
hv.extension('bokeh')
```

```python
# Read in expression data
df = pd.read_hdf('/mnt/data/expression/tcga_gtex_tpm_norm_filt.hd5')
# Subset
gtex = df[df.label == 'gtex'].sort_values('tissue')
normal = df[df.label == 'tcga-normal'].sort_values('tissue')
tumor = df[df.label == 'tcga-tumor'].sort_values('tissue')
ckcc = pd.read_hdf('/mnt/data/expression/ckcc.hd5')
# Read in drug genes
genes = df.columns[5:]
drug_genes = [x.split('\t')[0] for x in open('../data/druggable-genes.tsv', 'r').readlines()]
drug_genes = [x for x in drug_genes if x in df.columns]
```

```python
out_dir = '/mnt/outlier-runs/ckcc-run/'
```

# Candidates

```python
def plot_gene(sample_id, tissue, gene):
    # Define sample and tissue comparison
    sample = ckcc.loc[sample_id]
    tdf = gtex[gtex.tissue == tissue]
    # Plot
    plt.figure(figsize=(12, 4))
    _, upper = r.math.iqr_bounds(gtex[gene])
    plt.axvline(upper, label='GTEx-IQR-Cutoff', c='b')

    _, upper = r.math.iqr_bounds(tumor[gene])
    plt.axvline(upper, label='Tumor-IQR-Cutoff', c='orange')
    
    plt.axvline(sample[gene], label='N-of-1', c='r')
    sns.kdeplot(gtex[gene], label='GTEx', shade=True, alpha=0.5, c='b')
    sns.kdeplot(tumor[gene], label='Tumor', shade=True, alpha=0.5, c='orange')
    sns.kdeplot(tdf[gene], label=tissue, shade=True, alpha=0.5, c='g')
    # Grab P-value
    pvals = pd.read_csv(os.path.join(out_dir, sample_id, 'pvals.tsv'), sep='\t', index_col=0, header=None)
    plt.title(f'{gene} - P: {round(pvals.loc[gene][1], 4)}')
```
