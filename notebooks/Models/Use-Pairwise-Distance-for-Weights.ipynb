# Weights from MedianRank

Approximate weights via

$$ w = \frac{\sum Ranks}{Ranks}$$

$$ w = \frac{w}{\sum w}$$

```python
ranks = r.outlier.pairwise_distance_ranks(sample, gtex, genes, 'tissue')
training_group = gtex[gtex.tissue.isin(ranks.head(5).Group)]
```

```python
ranks.head()
```

```python
def weights_from_ranks(ranks, n=5):
    weights = ranks.head(n)
    weights = weights.MedianRank.sum() / weights.MedianRank
    return weights / weights.sum()
```

```python
weights = weights_from_ranks(ranks)
```

## Normalize MedianRank by Sample Count

```python
from sklearn.metrics import pairwise_distances

def pairwise_distance_ranks(sample, df, genes, group: str):
    dist = pairwise_distances(np.array(sample[genes]).reshape(1, -1), df[genes])
    dist = pd.DataFrame([dist.ravel(), df[group]]).T
    dist.columns = ['Distance', 'Group']
    dist = dist.sort_values('Distance')

    # Pandas-FU
    dist = dist.reset_index(drop=True).reset_index()
    dist = dist.groupby('Group').apply(lambda x: x['index'].median()).sort_values().reset_index(name='MedianRank')
    tissue_counts = [len(df[df[group] == x]) for x in dist.Group]
    dist['MedianRank-Norm'] = dist['MedianRank'] / tissue_counts
    return dist
```

```python
ranks = pairwise_distance_ranks(sample, gtex, genes, 'tissue')
ranks = ranks.sort_values('MedianRank-Norm')
```

```python
ranks.head()
```

```python
weights = ranks.head(5)
weights = weights['MedianRank-Norm'].sum() / weights['MedianRank-Norm']
weights = weights / weights.sum()
weights
```

# Define Model

```python
def run_model(sample,
              df,
              training_genes,
              weights,
              group: str = 'tissue',
              **kwargs):
    """
    Run Bayesian model using prefit Y's for each Gene and Dataset distribution

    Args:
        sample: N-of-1 sample to run
        df: Background dataframe to use in comparison
        training_genes: Genes to use during training
        group: Column to use to distinguish different groups
        **kwargs:

    Returns:
        Model and Trace from PyMC3
    """
    classes = sorted(df[group].unique())
    df = df[[group] + training_genes]

    # Collect fits
    ys = {}
    for gene in training_genes:
        for i, dataset in enumerate(classes):
            cat_mu, cat_sd = st.norm.fit(df[df[group] == dataset][gene])
            # Standard deviation can't be initialized to 0, so set to 0.1
            cat_sd = 0.1 if cat_sd == 0 else cat_sd
            ys[f'{gene}={dataset}'] = (cat_mu, cat_sd)
            
    print('Building model')
    with pm.Model() as model:
        # Linear model priors
        a = pm.Normal('a', mu=0, sd=1)
        # Model error
        eps = pm.InverseGamma('eps', 2.1, 1)

        # TODO: Try tt.stack to declare mu more intelligently via b * y
        # Linear model declaration
        for gene in tqdm(training_genes):
            mu = a
            for i, dataset in enumerate(classes):
                name = f'{gene}={dataset}'
                y = pm.Normal(name, *ys[name])
                mu += weights[i] * y

            # Embed mu in laplacian distribution
            pm.Laplace(gene, mu=mu, b=eps, observed=sample[gene])
        # Sample
        trace = pm.sample(**kwargs)
    return model, trace
```

```python
#training_genes = r.outlier.select_k_best_genes(training_group, genes
print('Calculating Beta Weights')
ranks = r.outlier.pairwise_distance_ranks(sample, gtex, genes, 'tissue')
weights = weights_from_ranks(ranks)
```
