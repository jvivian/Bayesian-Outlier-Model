# Mixture Model One Gene Example
Pick two tissues and build a mixture fit to both distributions


```python
from sklearn import mixture
```

```python
adrenal = gtex[gtex.tissue == 'Adrenal']
thyroid = gtex[gtex.tissue == 'Thyroid']
```

Subset thyroid to be same N as adrenal

```python
samples = np.random.choice(thyroid.index, len(adrenal))
thyroid = thyroid.loc[samples]
```

Plot gene expression

```python
gene = 'JAK1'
sns.kdeplot(adrenal[gene], label='Adrenal', shade=True, alpha=0.25)
sns.kdeplot(thyroid[gene], label='Thyroid', shade=True, alpha=0.25)
plt.title(f'Expression Distribution for {gene}')
plt.xlabel('Log2(TPM + 1)')
plt.ylabel('Density');
```

Plot combined expression

```python
dist = pd.concat([adrenal[gene], thyroid[gene]])
sns.kdeplot(dist, label='Adrenal+Thyroid', shade=True, alpha=0.5)
```

Try naive gaussian fit

```python
gmm = mixture.GaussianMixture(n_components=2).fit(pd.DataFrame(dist))
m1, m2 = gmm.means_
std1, std2 = gmm.covariances_
```

```python
x = np.linspace(min(dist), max(dist), len(dist))
plt.plot(x, *st.norm.pdf(x, m1, std1), label='u={}, o={}'.format(
    round(float(m1), 1), round(float(std1), 1))
)
plt.plot(x, *st.norm.pdf(x, m2, std2), label='u={}, o={}'.format(
    round(float(m2), 1), round(float(std2), 1))
)
plt.title(f'Mixture Model Fits for {gene}')
plt.xlabel('Log2(TPM + 1)')
plt.ylabel('Density');
```

Plot mixture distribution on top of merged distribution

```python
x, y = gmm.sample()
```

```python
x, y = gmm.sample(5000)

sns.kdeplot(dist, label='Adrenal+Thyroid', shade=True, alpha=0.5)
sns.kdeplot(x[:,0], label='Mixture', shade=True, alpha=0.5)
plt.title(f'Mixture and Composite Distributions for {gene}')
plt.xlabel('Log2(TPM + 1)')
plt.ylabel('Density');
```

# Create Mixture Sample

For every gene:
- Fit a mixture
- Sample one gene
    - Alternatively: Sample N times and take median

```python
sample_genes = []
for gene in tqdm(genes):
    dist = pd.concat([adrenal[gene], thyroid[gene]])
    gmm = mixture.GaussianMixture(n_components=2).fit(pd.DataFrame(dist))
    x, y = gmm.sample()
    sample_genes.append(float(x))
```

```python
mix_sample = pd.Series(sample_genes, index=genes)
mix_sample.name = 'Mix-Adrenal-Thyroid'
```

## Calculate Closest Tissues

```python
ranks = r.outlier.pca_distances(mix_sample, gtex, genes, 'tissue')
ranks.head()
```

That's exactly what we'd hope to see! Let's run the model now

# Run Model

```python
training_group = gtex[gtex['tissue'].isin(ranks.head(5).Group)]
    
training_genes = r.outlier.select_k_best_genes(training_group, genes, 'tissue', 40)
training_genes = list(set(drug_genes + training_genes))
```

```python
out_dir = '/mnt/research_serializations/Mixture-Model-Simulation-Validation'
```

```python
out = os.path.join(out_dir, 'mix-adrenal-thyroid.pkl')
if os.path.exists(out):
    m, t = r.outlier.load_model(out)
else:
    m, t = r.outlier.run_model(mix_sample, training_group, training_genes, 'tissue')
    r.outlier.pickle_model(out, m, t)
```
